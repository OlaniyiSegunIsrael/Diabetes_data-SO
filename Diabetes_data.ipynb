```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# Load the dataset
diabetes = load_diabetes()
data = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
data['target'] = diabetes.target

# Display first few rows
print(data.head())

# Describe the dataset
print(data.describe())
```

```python
# Create the figure
plt.figure(figsize=(10, 6))

# Positional arguments: x, y
# Keyword arguments: marker, linestyle, color
plt.plot(data['age'], data['target'], marker='o', linestyle='--', color='purple')

# Add title and axis labels
plt.title('Age vs Diabetes Progression')
plt.xlabel('Age (standardized)')
plt.ylabel('Target (disease progression)')

# Show the plot
plt.show()
```

```python
# Create the figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

#Subplot 1: BMI vs Target ---
ax1.scatter(data['bmi'], data['target'], color='blue', marker='o')
ax1.set_title('BMI vs Diabetes Progression')
ax1.set_xlabel('BMI (standardized)')
ax1.set_ylabel('Target (disease progression)')

#Subplot 2: BP vs Target ---
ax2.scatter(data['bp'], data['target'], color='red', marker='x')
ax2.set_title('Blood Pressure vs Diabetes Progression')
ax2.set_xlabel('Blood Pressure (standardized)')
ax2.set_ylabel('Target (disease progression)')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#Ticks, Labels, and Legends
# Create the figure
plt.figure(figsize=(10, 6))

# Scatter plot with customization
plt.scatter(data['s1'], data['target'], color='green', marker='^', label='S1 vs Target')

# Axis labels
plt.xlabel('S1 (standardized)', fontsize=12)
plt.ylabel('Target (disease progression)', fontsize=12)

# Ticks with fontsize 10
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Title
plt.title('S1 vs Diabetes Progression', fontsize=14)

# Legend and grid
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
#Colors, Markers, and Line Styles
import matplotlib.pyplot as plt

# Create the figure
plt.figure(figsize=(10, 6))

# Plot s1 vs target
plt.plot(data['s1'], data['target'], color='blue', marker='o', linestyle='-', label='S1 vs Target')

# Plot s2 vs target
plt.plot(data['s2'], data['target'], color='red', marker='s', linestyle='--', label='S2 vs Target')

# Plot s3 vs target
plt.plot(data['s3'], data['target'], color='green', marker='^', linestyle=':', label='S3 vs Target')

# Labels and title
plt.xlabel('Feature Value (standardized)')
plt.ylabel('Target (disease progression)')
plt.title('S1, S2, and S3 vs Diabetes Progression')

# Legend
plt.legend()

# Display the plot
plt.show()
```

```python
#Legends and grid
# Create the figure
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(data['s4'], data['target'], color='purple', marker='o', label='S4 vs Target')

# Labels and title
plt.xlabel('S4 (standardized)')
plt.ylabel('Target (disease progression)')
plt.title('S4 vs Diabetes Progression')

# Legend (loc=2 means upper-left corner)
plt.legend(loc=2)

# Grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
#Saving Plots as an Image File
# Create the figure
plt.figure(figsize=(10, 6))

#Plot age vs target with a marker
plt.plot(data['age'], data['target'], marker='o', linestyle='-', color='blue')

#Labels and title
plt.xlabel('Age (standardized)')
plt.ylabel('Target (disease progression)')
plt.title('Age vs Diabetes Progression')

#Save the figure
plt.savefig('age_vs_target.png')
plt.savefig('age_vs_target.pdf')

#Display the plot
plt.show()
```
